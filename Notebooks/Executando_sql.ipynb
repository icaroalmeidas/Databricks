```sql
--Criando banco de dados
CREATE DATABASE IF NOT EXISTS bike_store_bd;
```

```sql
--Criando tabela temporaria usando o arquivo csv de base
USE default;
CREATE OR REPLACE TEMPORARY VIEW products_tmp
using csv
OPTIONS(
  'path' '/Volumes/workspace/default/files-studies/Bike Store/products.csv',
  'header' 'true',
  'inferschema' 'true'
)
```

```sql
--Criando tabela no banco de dados usando a tabela temporaria.
create or replace table bike_store_bd.products as select * from products_tmp
```

```sql
--Verificando a tabela criada
Select * from bike_store_bd.products limit 10
```

```sql
--Verificando quantidade de produtos por ano
select 
  model_year as Ano,
  count(product_id) as QTD
  from bike_store_bd.products
  group by model_year
  order by QTD desc
```

```sql
--Criando tabela temporaria usando o arquivo csv de base
USE default;
CREATE OR REPLACE TEMPORARY VIEW customers_tmp
using csv
OPTIONS(
  'path' '/Volumes/workspace/default/files-studies/Bike Store/customers.csv',
  'header' 'true',
  'inferschema' 'true'
)
;
create or replace table bike_store_bd.customers as select * from customers_tmp
```
