## Criando dataframes atraves de Consultas em SQL

```python
%sql
--Verificando dados da tabela
select * from bike_store_bd.customers limit 1
```

```python
%python
#Tranformando consulta em DataFrame
df = spark.sql('''
select customer_id,
    concat(first_name,' ', last_name) as Nome_completo,
    phone as Telefone,
    email as Email,
    street as Rua,
    city as Cidade,
    state as Estado,
    zip_code as CEP
from bike_store_bd.customers as Clientes
where state = 'NY'
'''
)
```

```python
%python
df.show(10)
```

```python
%python
type(df)
```

```python
%python
#Criando diretorio 
dbutils.fs.mkdirs('/Volumes/workspace/default/files-studies/Diferentes-Formatos')
```

```python
%py
#Salvando dataframe no formato parquet
df.write.mode("overwrite").parquet("/Volumes/workspace/default/files-studies/Diferentes-Formatos/clientesNY.parquet")
```

```python
%python
display(dbutils.fs.ls('/Volumes/workspace/default/files-studies/Diferentes-Formatos'))
```

```python
%py
# Salvando dataframe em csv e json
df.write.option("demilimiter", ",").option("header", "true").mode("overwrite").csv("/Volumes/workspace/default/files-studies/Diferentes-Formatos/clientesNY.csv")

df.write.mode("overwrite").json("/Volumes/workspace/default/files-studies/Diferentes-Formatos/clientesNY.json")
```
